#### ステップ 2: 協調フィルタリングモデルの訓練  
Surprise ライブラリを使用して、協調フィルタリングモデルを訓練します。  
データをトレーニングセットとテストセットに分割します。  


```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse
import pandas as pd

# 購入履歴データを読み込み
purchase_data = pd.read_csv('purchase_history.csv')

# Surprise ライブラリ用のデータセットを作成
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(purchase_data[['customer_id', 'car_id', '満足度']], reader)

# データをトレーニングセットとテストセットに分割
trainset, testset = train_test_split(data, test_size=0.25)

# SVD（Singular Value Decomposition）を使った協調フィルタリングモデルの訓練
algo = SVD()
algo.fit(trainset)

# 希望価格の範囲に応じた推薦を行う関数
def svd_collaborative_recommend_with_budget_range(new_customer, algo, cars_df):
    # 顧客の希望価格の範囲を取得
    price_range = new_customer['価格']
    min_price = price_range *0.8
    max_price = price_range *1.1

    # 希望価格範囲内の車種に対する推薦スコアを計算
    car_ids = cars_df['id']
    predicted_ratings = []
    for car_id in car_ids:
        car_price = cars_df.loc[cars_df['id'] == car_id, '価格'].values[0]
        if min_price <= car_price <= max_price:
            predicted_rating = algo.predict(new_customer['id'], car_id).est
            predicted_ratings.append((car_id, predicted_rating))

    # トップ5の推薦車種を選択
    top_recommendations = sorted(predicted_ratings, key=lambda x: x[1], reverse=True)[:5]
    recommended_cars = cars_df[cars_df['id'].isin([car_id for car_id, _ in top_recommendations])]
    return recommended_cars
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split

# 顧客データと購入履歴データの読み込み
customers = pd.read_csv('customers.csv')
purchase_history = pd.read_csv('purchase_history.csv')
cars = pd.read_csv('cars.csv')

# SVDモデルの訓練
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(purchase_history[['customer_id', 'car_id', '満足度']], reader)
trainset = data.build_full_trainset()
algo = SVD()
algo.fit(trainset)



# 新規顧客と既存顧客間の類似性を計算する関数
def calculate_similarity(new_customer, customers_df):
    # 顧客特徴のベクトル化（例: 年齢、性別、職業など）
    # 注意: 実際のデータに適合するように調整する必要があります
    customer_features = np.array(customers_df[['年齢', '性別', '職業']])
    new_customer_features = np.array(new_customer[['年齢', '性別', '職業']])
    similarities = cosine_similarity(new_customer_features, customer_features)
    return similarities

# 類似した顧客が選んだ車種を推薦する関数
def recommend_based_on_similar_customers(new_customer, customers_df, purchase_history_df, cars_df, algo, top_n=5):
    similarities = calculate_similarity(new_customer, customers_df)
    most_similar_customer_idx = np.argmax(similarities)
    most_similar_customer_id = customers_df.iloc[most_similar_customer_idx]['id']

    # 顧客の希望価格範囲を取得
    min_price = new_customer['価格']*0.8
    max_price = new_customer['価格']*1.1

    # 類似顧客が購入した車種を抽出
    purchased_car_ids = purchase_history_df[purchase_history_df['customer_id'] == most_similar_customer_id]['car_id'].unique()

    # SVDを使ってスコアリングし、価格範囲内の車種のみを選択
    predicted_ratings = []
    for car_id in purchased_car_ids:
        car_price = cars_df.loc[cars_df['id'] == car_id, '価格'].values[0]
        if min_price <= car_price <= max_price:
            predicted_rating = algo.predict(new_customer['id'], car_id).est
            predicted_ratings.append((car_id, predicted_rating))

    top_recommendations = sorted(predicted_ratings, key=lambda x: x[1], reverse=True)[:top_n]
    recommended_car_ids = [car_id for car_id, _ in top_recommendations]

    # 推薦された車種を返す
    recommended_cars = cars_df[cars_df['id'].isin(recommended_car_ids)]
    return recommended_cars

# 新規顧客のデータ
# new_customer = pd.DataFrame({'id': [999], '年齢': [30], '性別': [1], '職業': [2], '希望価格最小': [200], '希望価格最大': [300]})

# 推薦を生成
#recommended_cars = recommend_based_on_similar_customers(new_customer, customers, purchase_history, cars, algo)
#print(recommended_cars)
```

#### ステップ 3: コンテンツベースフィルタリングの実装  
顧客の好みに基づいて車種を推薦するコンテンツベースフィルタリングを実装します。


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 車種の特徴ベクトルを作成
def create_car_features(cars_df):
    # 特徴として「タイプ、色、価格、新車」を組み合わせる
    cars_df['features'] = cars_df.apply(lambda x: f"{x['タイプ']} {x['色']}{x['価格']} {x['新車']}", axis=1)
    return cars_df

# TF-IDF ベクトルの作成
cars_df = create_car_features(cars_df)
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(cars_df['features'])

# 顧客の好みに基づいて車種を推薦
def content_based_recommend(new_customer, cars_df):
    # 顧客の好みの特徴ベクトルを作成
    customer_pref = f"{new_customer['タイプ']} {new_customer['色']} {new_customer['価格']} {new_customer['新車']}"
    customer_pref_vector = tfidf.transform([customer_pref])

    # 類似度スコアの計算
    cosine_sim = cosine_similarity(customer_pref_vector, tfidf_matrix)
    #print(cosine_sim)
    
    # スコアに基づいて車種を推薦
    top_matches = cosine_sim.argsort()[0][-5:][::-1]
    recommended_cars = cars_df.iloc[top_matches]
    return recommended_cars
```

#### ステップ 4: ハイブリッド推薦システムの実装  
協調フィルタリングとコンテンツベースフィルタリングの結果を組み合わせるハイブリッド推薦システムを実装します。  


```python
def hybrid_recommendation(customer, algo, cars_df,customers_df, purchase_history_df, top_n=5):
    content_recommendations = content_based_recommend(customer, cars_df)
    #collaborative_recommendations = collaborative_recommend(customer, algo, cars_df)
    #collaborative_recommendations =  svd_collaborative_recommend_with_budget_range(customer, algo, cars_df)
    collaborative_recommendations =  recommend_based_on_similar_customers(customer, customers_df, purchase_history_df, cars_df, algo, top_n)
    # 両方の方法から得られた推薦を組み合わせる（例: 重複排除）
    final_recommendations = pd.concat([content_recommendations, collaborative_recommendations]).drop_duplicates().head(top_n)

    return final_recommendations
```
